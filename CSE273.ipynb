Where in NumPy

```python
arr=np.array([10,-5,20,-3,15])
result=np.where(arr<0,0,arr)#(condition,replace with,name of array)
print(result)
```

```python
arr=np.array([1,2,3,4,5])
result=np.where(arr%2!=0,-1,arr)
print(result)
```

```python
marks=np.array([45,78,32,90,55])
status=np.where(marks>=50,1,0)
print(status)
```

```python
data=np.array([10,20,30,40,50])
mean=np.mean(data)
result=np.where(data>mean,mean,data)
print(result)
```

Apply Where Condition in 2D Array

```python
mat=np.array([[10,60,30],[80,20,90]])
result=np.where(mat>=50,1,0)
print(result)
```
